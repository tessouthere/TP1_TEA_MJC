# Scatterplot and Pairwise

```python
x=drugs['Satisfaction']
y=drugs['Effective']

plt.scatter(x, y,)
plt.title('Satisfcation of Drugs vs Effectivness')
plt.xlabel('Satisfaction of Drug')
plt.ylabel('Effectiveness of Drug')

plt.show()
```

```python
x=drugs['EaseOfUse']
y=drugs['Effective']

plt.scatter(x, y,)
plt.title('Ease of Use of Drugs vs Effectivness')
plt.xlabel('Ease Of Use of Drug')
plt.ylabel('Effectiveness of Drug')

plt.show()
```

```python
x=drugs['EaseOfUse']
y=drugs['Satisfaction']

plt.scatter(x, y,)
plt.title('Ease of Use of Drugs vs Satisfaction')
plt.xlabel('Ease Of Use of Drug')
plt.ylabel('Satisfaction of Drug')

plt.show()
```

```python
sns.pairplot(drugs, hue='Form')
```

# Histograms and Distributions

```python
plt.hist(drugs['Satisfaction'], 
         facecolor='lightgreen', 
         edgecolor='blue', 
         bins=10)
plt.title('Histogram of Satisfaction Ratings')
plt.show()
```

```python
plt.hist(drugs['Effective'], 
         facecolor='lightgreen', 
         edgecolor='blue', 
         bins=10)
plt.title('Histogram of Effectiveness Ratings')
plt.show()
```

```python
plt.hist(drugs['EaseOfUse'], 
         facecolor='lightgreen', 
         edgecolor='blue', 
         bins=10)
plt.title('Histogram of Ease of Use Ratings')
plt.show()
```

```python
plt.hist(drugs['Price'], 
         facecolor='lightgreen', 
         edgecolor='blue', 
         bins=100,
         range=(0, 2000))
# Calculating the mean
mean=drugs['Price'].mean()
median=drugs['Price'].median()
#ploting the mean
plt.title("Histogram of Drug Price between 0 and 2000 Dollars")

plt.axvline(mean, color='red')
plt.axvline(median, color='green')
plt.show()
```

```python
boxplot=sns.boxplot(x='Price',
                    data=drugs)
plt.title('Boxplot of Price Distribution')
```

```python
histplot=sns.histplot(data=drugs,
                      x='Price')
# Calculating the mean
mean=drugs['Price'].mean()
median=drugs['Price'].median()
#ploting the mean
plt.axvline(mean, color='red')
plt.axvline(median, color='green')
plt.show()
```

```python
plt.pie([70.66,24.53,4.09,0.73], 
        labels=['RX','OTC','RX/OTC','\r\r\n'],
        autopct='%.1f')

plt.axis('equal')
plt.title('Pie Chart Frequency Distribution of Types of Drugs')

plt.show()
```
